````python
from ultralytics import YOLO
MODEL_PATH = '../yolov10/yolov10n.pt'
model = YOLO(MODEL_PATH)
````

````python
model.info()
````

````python
Yaml_path = 'safety_helmet_dataset\data.yaml'
EPOCHS = 1
IMG_SIZE = 640
BATHCH_SIZE = 60
````

````python
model.train(data = Yaml_path,
            epochs = EPOCHS,
            batch = BATHCH_SIZE,
            imgsz = IMG_SIZE)
````
