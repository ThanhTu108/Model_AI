## Training

````python
from ultralytics import YOLO
MODEL_PATH = '../yolov10/yolov10n.pt'
model = YOLO(MODEL_PATH)
````

````python
model.info()
````

````python
Yaml_path = 'safety_helmet_dataset\data.yaml'
EPOCHS = 100
IMG_SIZE = 64
BATHCH_SIZE = 32
````

````python
model.train(data = Yaml_path,
            epochs = EPOCHS,
            batch = BATHCH_SIZE,
            imgsz = IMG_SIZE)
````

## Predicting


````python
import matplotlib.pyplot as plt
import numpy as np
import cv2
train_path = '../runs/detect/train32/weights/best.pt'
img_path = 'Img_test/helmet.jpg'

train_model = YOLO(train_path)

result = train_model.predict(
    source=img_path,
    imgsz = IMG_SIZE,
    conf = 0.1,
)

img_result = result[0].plot()
# print(result[0].plot())
# print(img_result.shape)

# print(type(img_result))
cv2.imwrite('output.jpg', img_result)
plt.imshow(img_result[:, :, ::-1])
plt.show()
````
